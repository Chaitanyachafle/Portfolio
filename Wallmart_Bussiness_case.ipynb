```python
print("Mean:",round(df["Purchase"].mean(),2))
print("Median:",df["Purchase"].median())
```

```python
# IQR = Q3 - Q1
Q1 = df['Purchase'].quantile(0.25)
Q3 = df['Purchase'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f'IQR Purchase :{IQR}')
print(f'Lower bound Purchase :{lower_bound}')
print(f'Upper bound Purchase :{upper_bound}')
```

```python
data_cleaned = df[(df['Purchase'] >= lower_bound) & (df['Purchase'] <= upper_bound)]
data_cleaned
```

```python
sns.boxplot(y=data_cleaned['Purchase'])
```

```python
average_married_spending=round(data_cleaned[data_cleaned['Marital_Status']==1]["Purchase"].mean(),2)
average_single_spending=round(data_cleaned[data_cleaned['Marital_Status']==0]["Purchase"].mean(),2)
print(f'Average married spending: {average_married_spending}')
print(f'Average Single spending: {average_single_spending}')
```

Conclusion: The analysis indicates differences in spending patterns between married and unmarried customers

```python
CI=0.95
alpha=1-CI
popmean=np.mean(data_cleaned["Purchase"])
sample_size=8000
std_error=np.std(purchase_mean_100)/np.sqrt(8000)
z1=norm.ppf(0.025)
z2=norm.ppf(1-0.025)
x1=round(popmean + z1 * np.std(purchase_mean_100),2)
x2=round(popmean + z2 *np.std(purchase_mean_100),2)
print(f"95% confidence interval : {x1},{x2}")

```

Confidence intervals and distribution of the mean of the expenses by female and male customers

```python
CI=0.95
alpha=1-CI
popmean=np.mean(data_cleaned[data_cleaned['Gender']=="F"]["Purchase"])
sample_size=8000
std_error=np.std(purchase_mean_100)/np.sqrt(8000)
z1=norm.ppf(0.025)
z2=norm.ppf(1-0.025)
x1=round(popmean + z1 * np.std(purchase_mean_100),2)
x2=round(popmean + z2 *np.std(purchase_mean_100),2)
average_Female_spending=round(data_cleaned[data_cleaned['Gender']=="F"]["Purchase"].mean(),2)
print(f"95% confidence interval-Female : {x1},{x2}")
print(f"The average female spending {average_Female_spending} fall in between 95% CI {x1} and {x2}")


```

```python
CI=0.95
alpha=1-CI
popmean=np.mean(data_cleaned[data_cleaned['Gender']=="M"]["Purchase"])
sample_size=8000
std_error=np.std(purchase_mean_100)/np.sqrt(8000)
z1=norm.ppf(0.025)
z2=norm.ppf(1-0.025)
x1=round(popmean + z1 * np.std(purchase_mean_100),2)
x2=round(popmean + z2 *np.std(purchase_mean_100),2)
average_male_spending=round(data_cleaned[data_cleaned['Marital_Status']==0]["Purchase"].mean(),2)
print(f"95% confidence interval-Male : {x1},{x2}")
print(f"The average male spending {average_male_spending} fall in between 95% CI {x1} and {x2}")
```
